```python

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from IPython.display import display

print("TensorFlow 版本:", tf.__version__)
```

```python
posts = pd.read_excel("用户微博总.xlsx")
users = pd.read_excel("用户信息.xlsx")

print("✅ 已成功读取 Excel 文件")
print("\n微博表列名:", list(posts.columns))
print("用户表列名:", list(users.columns))
```

```python
df = posts.merge(users, on="uid", how="inner")

print("合并后数据集大小:", df.shape)
display(df.head())
```

```python
TEXT_COL_CANDIDATES = [
    "文本内容",         
    "转发文本内容",      
    "content", "text", "微博内容", "weibo", "微博文本",
]

GENDER_COL_CANDIDATES = [
    "性别", "gender", "sex"
]

FANS_COL_CANDIDATES = [
    "粉丝数", "fans_count", "followers"
]

LOC_COL_CANDIDATES = [
    "位置", "location", "loc", "地区", "省份", "城市"
]

def find_first_existing(cols, candidates, name):
    for c in candidates:
        if c in cols:
            print(f"✅ 使用 {name} 列: {c}")
            return c
    raise ValueError(f"在列名 {cols} 中找不到 {name} 列，请根据实际列名修改 *_{name.upper()}_COL_CANDIDATES*。")

cols = df.columns
TEXT_COL = find_first_existing(cols, TEXT_COL_CANDIDATES, "文本")
GENDER_COL = find_first_existing(cols, GENDER_COL_CANDIDATES, "性别")
FANS_COL = find_first_existing(cols, FANS_COL_CANDIDATES, "粉丝数")
LOC_COL = find_first_existing(cols, LOC_COL_CANDIDATES, "位置")


df = df[[TEXT_COL, GENDER_COL, FANS_COL, LOC_COL]].dropna()
df = df.sample(frac=1.0, random_state=42).reset_index(drop=True)
print("清洗后样本数:", len(df))
display(df.head())
```

```python
gender_series = df[GENDER_COL]

if gender_series.dtype == object:
    gender_values = sorted(gender_series.unique().tolist())
    gender_to_id = {g: i for i, g in enumerate(gender_values)}
    print("性别映射:", gender_to_id)
    y_gender = gender_series.map(gender_to_id).astype("float32").values
else:
    # 如果本来就是 0/1 或整数
    y_gender = gender_series.astype("float32").values

y_gender = y_gender.reshape(-1, 1)


y_fans_raw = df[FANS_COL].astype("float32").values
y_fans = np.log1p(y_fans_raw)   # log(1 + x) 平滑长尾分布
y_fans = y_fans.reshape(-1, 1)


loc_encoder = LabelEncoder()
y_location = loc_encoder.fit_transform(df[LOC_COL].astype(str).values)
num_locations = len(loc_encoder.classes_)

print("位置类别数:", num_locations)
print("位置类别列表:", loc_encoder.classes_)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_words = 20000   # 只保留最常见的若干字符
maxlen = 100        # 每条微博最多保留多少字符，多余截断，不足填充

texts = df[TEXT_COL].astype(str).tolist()

tokenizer = Tokenizer(num_words=max_words, char_level=True, oov_token="<UNK>")
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)
word_index = tokenizer.word_index

print("实际字符/Token 数（包含 <UNK>）:", len(word_index))

x = pad_sequences(sequences, maxlen=maxlen, padding="post", truncating="post")
print("输入张量形状:", x.shape)
```

```python
x_train, x_test, y_gender_train, y_gender_test, y_fans_train, y_fans_test, y_loc_train, y_loc_test = train_test_split(
    x, y_gender, y_fans, y_location,
    test_size=0.2,
    random_state=42
)

print("训练集大小:", x_train.shape[0])
print("测试集大小:", x_test.shape[0])
```

```python
embedding_dim = 64
num_words = min(max_words, len(word_index) + 1)

inputs = keras.Input(shape=(maxlen,), dtype="int32", name="weibo_text")

x_embed = layers.Embedding(
    input_dim=num_words,
    output_dim=embedding_dim,
    name="embedding"
)(inputs)

x_conv = layers.Conv1D(128, 5, activation="relu", padding="same", name="conv1")(x_embed)
x_pool = layers.GlobalMaxPooling1D(name="global_max_pool")(x_conv)
x_dense = layers.Dense(128, activation="relu", name="shared_dense")(x_pool)
x_dense = layers.Dropout(0.5, name="dropout")(x_dense)

gender_output = layers.Dense(1, activation="sigmoid", name="gender")(x_dense)
fans_output = layers.Dense(1, activation="linear", name="fans_count")(x_dense)
location_output = layers.Dense(num_locations, activation="softmax", name="location")(x_dense)

model = keras.Model(
    inputs=inputs,
    outputs=[gender_output, fans_output, location_output],
    name="weibo_multi_output_model"
)

model.summary()
```

```python
model.compile(
    optimizer="adam",
    loss={
        "gender": "binary_crossentropy",
        "fans_count": "mse",
        "location": "sparse_categorical_crossentropy",
    },
    loss_weights={
        "gender": 1.0,
        "fans_count": 0.5,
        "location": 1.0,
    },
    metrics={
        "gender": ["accuracy"],
        "fans_count": ["mse", "mae"],
        "location": ["accuracy"],
    },
)

history = model.fit(
    x_train,
    {
        "gender": y_gender_train,
        "fans_count": y_fans_train,
        "location": y_loc_train,
    },
    validation_split=0.2,
    epochs=5,          
    batch_size=128,
)
```

```python

plt.figure()
plt.plot(history.history.get("gender_accuracy", []), label="gender_acc_train")
plt.plot(history.history.get("val_gender_accuracy", []), label="gender_acc_val")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Gender prediction accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
eval_results = model.evaluate(
    x_test,
    {
        "gender": y_gender_test,
        "fans_count": y_fans_test,
        "location": y_loc_test,
    },
    verbose=2,
)

print("\n评价结果列表：")
print(eval_results)
```

```python
num_show = min(5, x_test.shape[0])
sample_indices = np.random.choice(np.arange(x_test.shape[0]), size=num_show, replace=False)

sample_x = x_test[sample_indices]
sample_texts = [texts[i] for i in sample_indices]

pred_gender, pred_fans, pred_loc = model.predict(sample_x)

def decode_gender(prob):
    return 1 if prob >= 0.5 else 0

import numpy as np

for text, g, f_log, l in zip(sample_texts, pred_gender, pred_fans, pred_loc):
    g_scalar = float(np.squeeze(g)) 
    fans_log_scalar = float(np.squeeze(f_log))

    print("微博文本:", text)
    print("  预测性别标签 (0/1):", decode_gender(g_scalar))
    print("  预测粉丝数（约值）:", int(np.expm1(fans_log_scalar)))
    loc_id = int(np.argmax(l))
    print("  预测位置类别:", loc_encoder.inverse_transform([loc_id])[0])
    print("-" * 60)

```

- **数据预处理**  
  - 对微博内容进行字符级 Tokenizer 编码，并统一填充到固定长度；  
  - 以 user_id 为键将微博表和用户表连接起来，只保留文本、性别、粉丝数和位置四个字段；  
  - 性别转换为 0/1，粉丝数取 log1p 以减弱长尾分布影响，位置用 LabelEncoder 编码。

- **模型结构**  
  - 使用 Keras 函数式 API 搭建单输入、多输出的网络：`Embedding + Conv1D + GlobalMaxPooling + Dense` 作为共享主体；  
  - 在共享表征之后接三路输出头：性别（二分类）、粉丝数（回归）、位置（多分类）。

- **训练与评估**  
  - 使用 `adam` 优化器，分别采用 `binary_crossentropy / mse / sparse_categorical_crossentropy` 作为三个任务的损失函数；  
  - 通过 `loss_weights` 对各任务损失做简单平衡；  
  - 观察训练 / 验证集上的准确率与损失曲线，评估是否出现过拟合或欠拟合。

- **实际意义**  
  - 在真实应用中，可以基于微博内容自动推断用户画像（性别、所在地区、用户影响力大致等级等），用于推荐系统、广告投放或用户分析；  
  - 多任务学习共享了文本编码部分的参数，相比分别训练三个单独模型，可以更好地利用样本，提高泛化能力。

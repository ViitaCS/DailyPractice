# 题目1 ：探索Chipotle快餐店的销售数据

步骤1 导入必要的库

```python
import pandas as pd
```

步骤2 导入数据集 chipotle.tsv，并存为变量名chipo

```python
from pathlib import Path

def pick_existing_file(candidates):
    for file in candidates:
        if Path(file).exists():
            return file
    raise FileNotFoundError(f"没找到数据文件，请把以下任一文件放到当前目录：{candidates}")

chipotle_file = pick_existing_file([
    'chipotle.tsv',
    'Chipotle/chipotle.tsv'
])
chipo = pd.read_csv(chipotle_file, sep='	')
```

步骤3 查看前10行内容

```python
chipo.head(10)
```

步骤4 数据集中有多少个列(columns)

```python
chipo.shape[1]
```

步骤5 查看全部的列名称

```python
chipo.columns
```

步骤6 查看数据集的索引index

```python
chipo.index
```

步骤7 下单数最多商品(item)是什么?

```python
item_orders = (
    chipo.groupby('item_name', as_index=False)['quantity']
         .sum()
         .sort_values('quantity', ascending=False)
)
item_orders.head(1)
```

步骤8 在item_name这一列中，一共有多少种商品被下单？

```python
chipo['item_name'].nunique()
```

步骤9 在choice_description中，下单次数最多的商品是什么？

```python
choice_orders = (
    chipo.dropna(subset=['choice_description'])
         .groupby('choice_description', as_index=False)['quantity']
         .sum()
         .sort_values('quantity', ascending=False)
)
choice_orders.head(1)
```

步骤10 一共有多少商品被下单？

```python
chipo['quantity'].sum()
```

步骤11 将item_price转换为浮点数

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)
chipo['item_price'].dtype
```

步骤12 在该数据集对应的时期内，收入(revenue)是多少？

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
revenue
```

步骤13 在该数据集对应的时期内，一共有多少订单？

```python
chipo['order_id'].nunique()
```

步骤14 每一单(order)对应的平均总价是多少？

```python
order_revenue = (
    chipo.assign(revenue=chipo['quantity'] * chipo['item_price'])
         .groupby('order_id')['revenue']
         .sum()
)
order_revenue.mean()
```

步骤15 一共有多少种不同的商品被售出？

```python
chipo['item_name'].nunique()
```

# 题目 3 ：探索1960 - 2014 美国犯罪数据

步骤1 导入必要的库

```python
import numpy as np
import pandas as pd
```

步骤2 导入数据集 US_Crime_Rates_1960_2014.csv，并存为变量名 crime

```python
from pathlib import Path

def pick_existing_file(candidates):
    for file in candidates:
        if Path(file).exists():
            return file
    raise FileNotFoundError(f"没找到数据文件，请把以下任一文件放到当前目录：{candidates}")

crime_file = pick_existing_file([
    'US_Crime_Rates_1960_2014.csv',
    'US_Crime_Rates.csv'
])
crime = pd.read_csv(crime_file)
```

步骤3 查看每一列(column)的数据类型是什么？

```python
crime.dtypes
```

步骤4 将Year的数据类型转换为 datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
crime.dtypes
```

步骤5 将列Year设置为数据框crime的索引index

```python
crime = crime.set_index('Year')
crime.head()
```

步骤6 删除名为Total的列

```python
crime = crime.drop(columns=['Total'])
crime.head()
```

步骤7 按照Year对数据框进行分组并求和

```python
# 这里按“年代”统计，便于回答下一问“哪个年代最危险”
decade_crime = crime.resample('10YS').sum(numeric_only=True)
# Population（人口）不应该直接按十年累加，这里保留每个年代最后一年的值
decade_crime['Population'] = crime['Population'].resample('10YS').max()
decade_crime.index = decade_crime.index.year
decade_crime
```

步骤8 何时是美国历史上生存最危险的年代？ 

```python
complete_decade_crime = decade_crime.loc[decade_crime.index < 2010]
most_dangerous_decade = complete_decade_crime['Violent'].idxmax()
f'最危险的年代大致是 {most_dangerous_decade}s'
```
